```python
df5=pd.DataFrame({
"country":["USA","USA","INDIA","INDIA"],
    "year":[2020,2021,2020,2021],
    "sales":[100,120,90,200],
    "profit":[20,25,18,22]
})
```

```python
df5
```

```python
df5.melt(
id_vars=["country","year"],
    value_vars=["sales","profit"],
    var_name="metrics",
    value_name="value"    
)
```

```python
melted_data=df5.melt(
id_vars=["country","year"],
    value_vars=["sales","profit"],
    var_name="metrics",
    value_name="value"    
)
```

```python
original=melted_data.pivot(
index=["country","year"],
    columns="metrics",
    values="value"
)
```

```python
#merge and joining of data
```

```python
df_customers=pd.DataFrame({
"customer_id":[1,2,3,4],
"name":["adam","bob","charlie","dave"]    
})
```

```python
df_order=pd.DataFrame({
    "order_id":[101,102,103,104],
    "customer_id":[2,1,4,5],
    "amount":[250,300,210,180]
})
```

```python
pd.merge(df_customers,df_order,on="customer_id")
```

```python
pd.merge(df_customers,df_order,on="customer_id",how="left")
```

```python
pd.merge(df_customers,df_order,on="customer_id",how="right")
```

```python
pd.merge(df_customers,df_order,on="customer_id",how="outer")
```
